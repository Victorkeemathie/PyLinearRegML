## Linear Regression Types

__a. Simple Linear Regression:__  
In simple linear regression, there is only one independent variable used to predict the dependent variable.  
It involves fitting a straight line to the data points to represent the relationship between the variables.  
The equation of a simple linear regression model is typically represented as: ` $Y = β0 + β1X + ε$ `,  
where ` $Y$ ` is the dependent variable, ` $X$ `is the independent variable, ` $β0$ ` is the intercept, ` $β1$ ` is the coefficient (slope), and ε is the error term.  
***  
__b. Multiple Linear Regression:__  
Multiple linear regression involves more than one independent variable to predict the dependent variable.  
It extends the concept of simple linear regression to multiple dimensions.  
The equation of a multiple linear regression model is: ` $Y = β0 + β1X1 + β2X2 + ... + βnXn + ε$ `,  
where Y is the dependent variable,  ` $X1$, $X2$, ..., $Xn$ ` are the independent variables, ` $β0$ ` is the intercept, ` $β1$, $β2$, ..., $βn$ ` are the coefficients, and $ε$  is the error term.  
***  


## __Assessing the Performance of a Linear Regression Model__

**Mean Squared Error (MSE):**

The mean squared error calculates the average squared difference between the predicted values and the actual values. It penalizes larger errors more heavily. 

$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) 
^2$$

where $n$ is the number of observations, $y_i$ is the actual value, and $\hat{y}_i$ is the predicted 
value.

**Root Mean Squared Error (RMSE):**

RMSE is the square root of the mean squared error and provides a measure of the average magnitude of errors.

$$RMSE = \sqrt{MSE}$$

Like MSE, lower RMSE values indicate better model performance.

**Mean Absolute Error (MAE):**

MAE calculates the average absolute difference between the predicted and actual values.

$$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

MAE is less sensitive to outliers compared to MSE.

**Coefficient of Determination ($R^2$ or R-squared):**

$R^2$ measures the proportion of the variance in the dependent variable that is predictable from the independent variables. It ranges from 0 to 1, where 1 indicates perfect prediction.

$$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$

where $\bar{y}$ is the mean of the dependent variable.

**Adjusted $R^2$:**

Adjusted $R^2$ is an extension of $R^2$ that penalizes the addition of irrelevant variables to the model. It accounts for the number of predictors in the model.

$$\text{Adjusted } R^2 = 1 - \frac{(1 - R^2)(n - 1)}{n - p - 1}$$

where $p$ is the number of predictors.

**OVERFITTING, UNDERFITTING, AND THE BIAS-VARIANCE TRADE-OFF IN MACHINE LEARNING**

**Overfitting:**

- Occurs when a model becomes too closely aligned with the training data, capturing noise and irrelevant patterns.
- Results in high performance on the training set but poor generalization to new, unseen data.
- Characterized by **low bias** and **high variance**.

**Underfitting:**

- Occurs when a model fails to capture the underlying patterns and relationships in the data.
- Results in poor performance on both the training and testing sets.
- Characterized by **high bias** and **low variance**.

**Bias-Variance Trade-off:**

- A fundamental concept in machine learning that describes the interplay between bias and variance.
- Involves three components:
    - **Bias Error:**
        - Represents the difference between the expected model prediction and the true value.
        - High bias suggests oversimplification of the target function.
        - Low bias suggests capturing more complexities of the target function.
    - **Variance Error:**
        - Represents the variability of model predictions across different training sets.
        - High variance suggests sensitivity to specific training data.
        - Low variance suggests predictions remain consistent across different data.
    - **Irreducible Error:**
        - Represents noise inherent in the data that cannot be eliminated.

**Goal:**

- To achieve both **low bias** and **low variance** for optimal model performance.
- Balancing these components requires careful consideration of model complexity, training data, and algorithm selection.

**Key Points:**

- Overfitting captures noise, leading to high variance and poor generalization.
- Underfitting misses important patterns, leading to high bias and inaccurate predictions.
- The bias-variance trade-off highlights the delicate balance between model complexity and generalization ability.
- Finding the right balance depends on the specific problem and dataset.

**Strategies to Combat Overfitting:**

- Regularization: Penalizes model complexity to reduce variance.
- Cross-validation: Evaluates model performance on different data subsets to assess generalization.
- Data augmentation: Increases training data diversity to reduce model sensitivity to specific data quirks.
- Early stopping: Terminates training when validation performance starts to degrade, preventing overfitting.

**Strategies to Combat Underfitting:**

- Increase model complexity (e.g., add more features or layers).
- Collect more training data.
- Feature engineering: Create more informative features.

**Understanding and managing the bias-variance trade-off is essential for building robust and generalizable machine learning models.**

### __Plotting Learning Curves__

```python
from sklearn.model_selection import learning_curve, ShuffleSplit
```

```python
def plot_learning_curve(estimator, title, X,y, ylim=None, cv=None,
                        train_size=np.linspace(0.1, 1, 10)):
    plt.figure()
    plt.title(title)
    plt.xlabel('Training Examples')
    plt.ylabel('Score')
    
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, train_sizes=train_size)
    
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color='red')
    
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, color='green')
    
    plt.plot(train_sizes, train_scores_mean, 'o-', color='red', label='Training Score')
    plt.plot(train_sizes, test_scores_mean, 'o-', color='green', label='Test Score')
    
    plt.legend(loc='best')
    
    return plt
    
```

```python
title = 'Learning Curve for Linear Regression'
cv = ShuffleSplit(n_splits=100, test_size=0.2, random_state=0)
model = LinearRegression()
plot_learning_curve(model, title,  X, y, ylim=(0.7, 1.01), cv=cv)
plt.show()
```
